```python
import pandas as pd

twitter_accounts = pd.read_csv('congress_twitter_users.csv')

twitter_id_to_party = {}

for index, row in twitter_accounts.iterrows():
    twitter_id_to_party[row['id']] = {
            'party':row['party'],
            'name': row['name'],
            'screen_name':row['screen_name']
        }
    

# twitter_accounts

#supported twitter ids to mine from 

ids = set([id for id in twitter_id_to_party.keys()])
ids
```
