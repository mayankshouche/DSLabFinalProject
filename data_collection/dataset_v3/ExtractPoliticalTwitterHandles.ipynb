```python
types = set()
for i in users:
    types.add(i['type'])
types
```

```python
caucus = None
committe = None
member = None
party = None

twitter_accounts = []
#     party = ''
#     type = ''
#     name = ''
#     id = ''
#     screen_name = ''
for i in users:
    if i['type'] == 'caucus':
        if i['party'] == 'N/A':
            continue
        party = i['party']
        for account in i['accounts']:
            account_builder = {
                'type':i['type'],
                'name':i['name'],
                'screen_name':'',
                'party':party,
                'id':''
            }
            account_builder['id'] = account['id']
            account_builder['screen_name'] =account['screen_name']
            twitter_accounts.append(account_builder)

    elif i['type'] == 'committee':
        for account in i['accounts']:
            account_builder = {
                'type':i['type'],
                'name':i['name'],
                'screen_name':'',
                'party':'',
                'id':''
            }
            if 'party' not in account or account['party'] == 'N/A':
                continue
            account_builder['id'] = account['id']
            account_builder['screen_name'] =account['screen_name']
            account_builder['party'] = account['party']
            twitter_accounts.append(account_builder)
    elif i['type'] == 'member':
        if i['party']=='N/A':
            continue

        for account in i['accounts']:
            if account['account_type']=='office':
                account_builder = {
                    'type':i['type'],
                    'name':i['name'],
                    'screen_name':'',
                    'party':i['party'],
                    'id':''
                }
                account_builder['id'] = account['id']
                account_builder['screen_name'] =account['screen_name']
                twitter_accounts.append(account_builder)
    elif i['type'] == 'party':
        for account in i['accounts']:
            account_builder = {
                'type':i['type'],
                'name':i['name'],
                'screen_name':'',
                'party':i['party'],
                'id':''
            }
            account_builder['id'] = account['id']
            account_builder['screen_name'] =account['screen_name']
            twitter_accounts.append(account_builder)
        
```

```python
twitter_accounts
```

```python
import csv
csv_file = 'congress_twitter_users.csv'


try:
    with open(csv_file, 'w') as csvfile:
        csv_columns = ['type', 'name', 'screen_name', 'party', 'id']
        writer = csv.DictWriter(csvfile, fieldnames=csv_columns)
        writer.writeheader()
        for account in twitter_accounts:
            writer.writerow(account)
except IOError:
    print("I/O error")
```
